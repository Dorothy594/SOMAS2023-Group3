```python
import matplotlib.pyplot as plt
import numpy as np

# 给定的数据点
points = [(2, 1), (2, 4), (5, 1), (7, 6)]

# 提取点的坐标
x_values, y_values = zip(*points)

# 绘制数据点
plt.scatter(x_values, y_values, color='blue', label='Data Points')

# 绘制直线x-y-1=0
x_line = np.linspace(min(x_values), max(x_values), 100)
y_line = x_line - 1
plt.plot(x_line, y_line, color='red', label='Line x-y-1=0')

# 计算并绘制点到直线的投影
for x, y in zip(x_values, y_values):
    projection_x = (x + y + 1) / 2  # 投影点的 x 坐标
    projection_y = projection_x - 1  # 投影点的 y 坐标
    plt.plot([x, projection_x], [y, projection_y], color='green', linestyle='--')  # 绘制投影线
    plt.scatter(projection_x, projection_y, color='green', marker='o')  # 绘制投影点

# 设置坐标轴标签
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 给定的数据点
points = [(1/np.sqrt(2), 3/np.sqrt(2)), (-2/np.sqrt(2), 6/np.sqrt(2)), (4/np.sqrt(2), 6/np.sqrt(2)), (1/np.sqrt(2), 13/np.sqrt(2))]

# 提取点的坐标
x_values, y_values = zip(*points)

# 绘制数据点
plt.scatter(x_values, y_values, color='blue', label='Data Points')

# 绘制直线x=0
plt.axvline(x = 1/np.sqrt(2), color='red', linestyle='--', label='Line x=1/np.sqrt(2)')

# 计算并绘制点到直线的投影
for x, y in zip(x_values, y_values):
    plt.plot([x, 1/np.sqrt(2)], [y, y], color='green', linestyle='--')  # 绘制投影线
    plt.scatter(1/np.sqrt(2), y, color='green', marker='o')  # 绘制投影点

# 设置坐标轴标签
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import math

class ml_3e(nn.Module):
    def __init__(self):
        super(ml_3e, self).__init__() # should be ml_3e, but suit the latex format
        input_dim = 3
        hidden_dim = 3
        output_dim = 2

        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, output_dim)
    
    def forward(self, x):
        x = torch.relu(self.fc1(x))  
        x = self.fc2(x)
        return x

model = ml_3e()

# input 
# non-linear transformation: (x, x^2+2, e(x))
inputs = torch.tensor([[1.0, 3.0, math.e], [2.0, 6.0, math.exp(2)]], dtype=torch.float32)
targets = torch.tensor([[1.0, 5.0], [1.0, 7.0]], dtype=torch.float32)

criterion = nn.MSELoss()  # using mse loss
optimizer = optim.SGD(model.parameters(), lr=0.01)  # optimizer using SGD

# train
for epoch in range(1000):  
    optimizer.zero_grad()
    outputs = model(inputs)
    loss = criterion(outputs, targets)
    loss.backward()
    optimizer.step()

# test
test_inputs = torch.tensor([[1.0, 3.0, math.e], [2.0, 6.0, math.exp(2)]], dtype=torch.float32)
predictions = model(test_inputs)
print(predictions)

# print weights
for name, param in model.named_parameters():
    if param.requires_grad:
        print(name, param.data)
```
